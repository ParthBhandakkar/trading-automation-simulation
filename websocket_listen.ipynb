```python
import websocket
import json
import nest_asyncio

nest_asyncio.apply()

# Function to handle incoming messages
def on_message(ws, message):
    try:
        # Attempt to parse the message as JSON
        data = json.loads(message)
        print(f"Received data: {json.dumps(data, indent=4)}")  # Pretty print the JSON data
    except json.JSONDecodeError:
        # If the message is not JSON, just print it as a string
        print(f"Received non-JSON message: {message}")

# Function to handle connection opening
def on_open(ws):
    print("Connected to WebSocket server")

# Function to handle errors
def on_error(ws, error):
    print(f"Error: {error}")

# Function to handle connection closing
def on_close(ws, close_status_code, close_reason):
    print("Connection closed")
    print(f"Close status code: {close_status_code}, reason: {close_reason}")

# WebSocket URL (replace with your WebSocket server URL)
websocket_url = "ws://localhost:12345"  # Replace with actual WebSocket server URL

# Create WebSocket connection and specify event handlers
ws = websocket.WebSocketApp(
    websocket_url,
    on_open=on_open,
    on_message=on_message,
    on_error=on_error,
    on_close=on_close
)

# Run the WebSocket client (this will keep the connection open and listen for messages)
ws.run_forever()
```
